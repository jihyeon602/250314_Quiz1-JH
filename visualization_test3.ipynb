```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

file_path = 'c:\\Users\\302-26\\jh\\OTg6QzA_activities.csv'
df = pd.read_csv(file_path)

# timestamp를 datetime 형식으로 변환
df['timestamp'] = pd.to_datetime(df['timestamp'])
print(df.dtypes)
print(df) #66281, 4
```

#dt 속성 접근자
- pandas의 datetime 속성 접근자
- 주요 속성 및 메서드:
day_name(): 요일 이름을 반환합니다.


hour: 시간 정보 반환

minute: 분 정보 반환

second: 초 정보 반환

year: 연도 정보 반환

month: 월 정보 반환

day: 일 정보 반환

```python
# 요일, 시간 추출
df['day_of_week'] = df['timestamp'].dt.day_name()
df['hour'] = df['timestamp'].dt.hour

#timestamp를 인덱스로 설정
df.set_index('timestamp', inplace=True)
print(df)
```

```python
df['p_x'] = df['X'].shift(1)
df['p_y'] = df['Y'].shift(1)
df['p_z'] = df['Z'].shift(1)

df.dropna(inplace=True)

def calc_dist(a):
    x_d = a['X'] - a['p_x']
    y_d = a['Y'] - a['p_y']
    z_d = a['Z'] - a['p_z']
    return np.sqrt(x_d**2 + y_d**2 + z_d**2)

df['dist'] = df.apply(calc_dist, axis=1)
df.head(5)
```
